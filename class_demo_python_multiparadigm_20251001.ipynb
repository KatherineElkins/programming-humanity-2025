## **(a) Imperative Programming Style**
It executes commands in a step-by-step manner, just like a series of verbal commands. Following the "how-to-solve" approach, it makes direct changes to the state of the program; hence it is also called the stateful programming model

## **(b) Procedural Programming Style**

The procedural programming paradigm is a subtype of imperative programming in which statements are structured into procedures (also known as subroutines or functions). The program composition is more of a procedure call where the programs might reside somewhere in the universe and the execution is sequential, thus becoming a bottleneck for resource utilization.

Like the imperative programming paradigm, procedural programming follows the stateful model. The procedural programming paradigm facilitates the practice of good program design and allows modules to be reused in the form of code libraries.

````
# NOTE: The code in the article does not work as is
#       Here is one implementation, not the most efficient

# imports and var declarations
sample_characters = ['p','y','t','h','o','n']

# Procedural Style
def stringify(characters):
  string_lst = []
  string_str = ''
  for c in characters:
    string_lst.append(c)
    print(''.join(string_lst))
  string_str = ''.join(string_lst)
  return string_str

# Call Procedure 'stringify)
# stringified_str = stringify(sample_characters)
# print(stringified_str)
print(stringify(sample_characters))
````

## **(c)Object Oriented Programming Style**

The object-oriented programming paradigm considers basic entities as objects whose instance can contain both data and the corresponding methods to modify that data. The different principles of object-oriented design help code reusability, data hiding, etc., but it is a complex beast, and writing the same logic an in object-oriented method is tricky. For example:

````
# imports and var declarations
sample_characters = ['p','y','t','h','o','n']

# OOP Style
class StringOps:
  def __init__(self, characters):
    self.characters = characters
  def stringify(self):
    self.string = ''.join(self.characters)

# Instantiate Object from Class
sample_string = StringOps(sample_characters)

# Call method 'stringify()' on object 'sample_string'
sample_string.stringify()

# Access attribute/data 'string' within the object 'sample_string'
sample_string.string
````

````
# class Emp has been defined here
class Emp:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def info(self):
        print("Hello, % s. You are % s old." % (self.name, self.age))

# Objects of class Emp has been
# made here
Emps = [Emp("John", 43),
    Emp("Hilbert", 16),
    Emp("Alice", 30)]

# Objects of class Emp has been
# used here
for emp in Emps:
    emp.info()
````

## **(d) Functional Programming Style**

The functional programming paradigm treats program computation as the evaluation of mathematical functions based on lambda calculus. Lambda calculus is a formal system in mathematical logic for expressing computation based on function abstraction and application using variable binding and substitution. It follows the "what-to-solve" approachâ€”that is, it expresses logic without describing its control flowâ€”hence it is also classified as the declarative programming model.

The functional programming paradigm promotes stateless functions, but it's important to note that Python's implementation of functional programming deviates from standard implementation. Python is said to be an impure functional language because it is possible to maintain state and create side effects if you are not careful. That said, functional programming is handy for parallel processing and is super-efficient for tasks requiring recursion and concurrent execution.

````
# imports and var declarations
import functools
sample_characters = ['p','y','t','h','o','n']

# Functional Style
sample_string = functools.reduce(lambda s,c: s + c, sample_characters)

# Pring out result
sample_string
````

# **(e) Logic Programming**



### ðŸ§  Logic Programming Explained: Building a Knowledge Base

This section explains the concept of **Logic Programming** using the `pyDatalog` example above.

If you're a first-time programmer, you're probably used to **Imperative Programming**â€”writing step-by-step instructions (e.g., "First, get the number. Second, add 5. Third, print the result.").

Logic Programming is different. It's **Declarative**. Instead of telling the computer *how* to solve a problem, you tell it *what* is true, and it figures out the solution itself.

---

### The Family Tree Analogy

Think of Logic Programming as building a knowledge base, like a family tree puzzle. You provide two things:

1.  **Facts:** The basic, undeniable truths (e.g., "Homer is a parent of Bart").
2.  **Rules:** General statements that allow for deduction (e.g., "A grandparent is a parent of a parent").

Once the knowledge base is built, you can ask **Queries** (questions), and the computer will use its facts and rules to find the answer.

### 1. Facts (The Data)

In the code, **Facts** are the basic truths we assert using the `+` operator:

```python
+ parent('Homer', 'Bart')
+ parent('Abe', 'Homer')
```

*   **What it means:** We are telling the computer that the relationship `parent` holds true for the pair `('Homer', 'Bart')` and `('Abe', 'Homer')`.
*   **In simple terms:** This is the raw data we feed into the system.

### 2. Rules (The Logic)

**Rules** are general statements that allow the computer to deduce new information. This is where the "logic" comes in.

```python
grandparent(X, Z) <= parent(X, Y) & parent(Y, Z)
```

*   **Variables (X, Y, Z):** These are placeholders, like unknowns in an algebra problem. They must start with an uppercase letter.
*   **The `&` symbol:** This means **AND**.
*   **The `<=` symbol:** This means **IF**.

*   **What it means:** "X is a grandparent of Z **IF** X is a parent of some person Y **AND** that person Y is a parent of Z."

The computer doesn't need to be told *how* to find a grandparent; it just needs the rule, and it will search through all its facts to find names that fit the placeholders (X, Y, and Z).

### 3. Queries (The Questions)

A **Query** is simply asking the system to find all the values that make a statement true.

```python
# Query: Find X such that parent(X, 'Bart') is true.
print(parent(X, 'Bart').data)
```

*   **What it means:** "Who is the parent (X) of Bart?"
*   **The result:** `[('Homer',)]`â€”The computer found that if it substitutes 'Homer' for X, the fact `parent('Homer', 'Bart')` is true.

```python
# Query: Find X such that grandparent(X, 'Bart') is true.
print(grandparent(X, 'Bart').data)
```

*   **What it means:** "Who is the grandparent (X) of Bart?"
*   **The process:** The computer uses the `grandparent` rule, finds that `Y` can be 'Homer', and then finds that `X` can be 'Abe', thus deducing the new fact: `grandparent('Abe', 'Bart')`.
*   **The result:** `[('Abe',)]`

## Logical Relations: Definitions

````
# 1. Define the terms (predicates and variables)
# Variables must start with an uppercase letter (X, Y, Z)
# Predicates (relations) must start with a lowercase letter (parent, grandparent)
pyDatalog.create_terms('parent, grandparent, X, Y, Z')
````

````
# 2. Assert the facts
# The '+' operator asserts a fact. Constants (names) must be quoted strings.
+ parent('Homer', 'Bart')
+ parent('Homer', 'Lisa')
+ parent('Abe', 'Homer')
````

````
# 3. Define the rule: Grandparent(X, Z) if Parent(X, Y) AND Parent(Y, Z)
# The '<=' operator means 'if'
# The '&' operator means 'and'
grandparent(X, Z) <= parent(X, Y) & parent(Y, Z)
````

## Logical Queries

````
# Query 1: Who is Bart's parent? (Find X such that parent(X, 'Bart'))
print("Who is Bart's parent?")
# The .data attribute returns the results as a list of tuples
print(parent(X, 'Bart').data)
````

````
# Query 2: Who is Homer's child? (Find X such that parent('Homer', X))
print("\nWho are Homer's children?")
print(parent('Homer', X).data)
````

````
# Query 3: Who is Bart's grandparent? (Find X such that grandparent(X, 'Bart'))
print("\nWho is Bart's grandparent?")
print(grandparent(X, 'Bart').data)
````
